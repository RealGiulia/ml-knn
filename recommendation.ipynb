```python
def notas_do_usuario(usuario):
    notas_do_usuario = notas.query('UsuarioID==%d' %usuario)
    notas_do_usuario = notas_do_usuario[['FilmeId', 'Notas']].set_index('FilmeId')
    return notas_do_usuario
```

```python
def distancia_usuario(usuario_Id1, usuario_Id2):
    notas1 = notas_do_usuario(usuario_Id1)
    notas2 = notas_do_usuario(usuario_Id2)
    diferencas = notas1.join(notas2, lsuffix='_esquerda', rsuffix='_direita').dropna()
    distancia = distancia_vetores(diferencas['Notas_esquerda'], diferencas['Notas_direita'])
    return[usuario_Id1, usuario_Id2, distancia]
```

```python
def distancia_de_todos(voce_Id, numero_de_usuarios_a_analisar = None):
    all_users = notas['UsuarioID'].unique()
    if numero_de_usuarios_a_analisar:
        all_users = all_users[:numero_de_usuarios_a_analisar]
    distancia = [distancia_usuario(voce_Id, usuario_Id) for usuario_Id in all_users]
    distancia = list(filter(None, distancia))
    distancia = pd.DataFrame(distancia, columns = ["voce", "outra_pessoa", "distancia"])
    return distancia
```

```python
voce = 1
def sugere_para(voce, numero_de_usuarios_a_analisar= None): 
    notas_de_voce = notas_do_usuario(voce)
    filmes_que_vc_ja_viu = notas_de_voce.index

    similares = mais_proximos_de(1, numero_de_usuarios_a_analisar)
    similar = similares.iloc[0].name
    notas_do_similar = notas_do_usuario(similar)
    notas_do_similar = notas_do_similar.drop(filmes_que_vc_ja_viu, errors = 'ignore')
    recomendacoes = notas_do_similar.sort_values("Notas", ascending=False)
    return recomendacoes.join(filmes)
```
